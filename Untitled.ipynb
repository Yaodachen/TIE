```python
def maopao(lst):
    for i in range(len(lst)):
        x = lst[i]
        for j in range(1,len(lst)-i):
            if lst[j]<lst[j-1]:
                lst[j-1],lst[j] = lst[j],lst[j-1]
    return(lst)
```

```python
def charu(lst):
    for i in range(1,len(lst)):
        x = lst[i]
        j = i
        while j > 0 and lst[j-1]>x:
            lst[j],lst[j-1] = lst[j-1],lst[j]
            j -= 1
    return(lst)
            
            
```

```python
def qksort(lst, l, r):
    pivot = lst[l]
    i = l
    j = r
    while i<j:
        while j>i and lst[j]>pivot:
            j -= 1
        if i<j:
            lst[i] = lst[j]
            i += 1
        while i<j and lst[i]<pivot:
            i += 1
        if i<j:
            lst[j] = lst[i]
            j -= 1
    lst[i] = pivot
    if l<i-1:
        lst = qksort(lst,l,i-1)
    if i+1<r:
        lst = qksort(lst,i+1,r)
    return lst
        
            
            
```
